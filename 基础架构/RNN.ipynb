# RNN

```python
input_size = 10
hidden_size = 20
batch_size = 2
torch.manual_seed(21)
rnn = nn.RNN(10, 20 ,2, batch_first=True)
rnn
```

```python
input = torch.randn(1, 3, 10)
input
```

```python
h0 = torch.randn(2, 1, 20)
h0
```

# RNNCell

```python
input_size = 10
hidden_size = 20
batch_size = 2
torch.manual_seed(21)
rnn_cell = nn.RNNCell(input_size, hidden_size)
```

```python
input_cell = torch.randn(batch_size, input_size)
input_cell
```

```python
hc0 = torch.randn(batch_size, hidden_size)
hc0
```

```python
out = rnn_cell(input_cell, hc0)
```

```python
out
```

```python
input = torch.randn(6, 3, 10) #每一个batch代表一个序列在时间t的token
input
```

```python
input[0]
```

```python
hx = torch.randn(3, 20)
hx
```

```python
output = []
for i in range(6):
    hx = rnn_cell(input[i], hx)
    output.append(hx)
```

```python
output #最后一个是三个序列的final hidden state
```
